```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib.ticker import PercentFormatter

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
with open('nobel.csv','r') as f:
    Data = pd.read_csv(f)
    
    Time = pd.to_datetime(Data["birth_date"]).dt.year
    Data['Time'] =  Data['year'] - Time     

    
    Chemistry = Data['Time'][Data['category']=='Chemistry']
    Literature = Data['Time'][Data['category']=='Literature']
    Medicine = Data['Time'][Data['category']=='Medicine']
    Peace = Data['Time'][Data['category']=='Peace']
    Physics = Data['Time'][Data['category']=='Physics']
    Economics = Data['Time'][Data['category']=='Economics']
    
    
    Data['Age_che']= Chemistry
    Data['Age_Liter']= Literature
    Data['Age_Med']= Medicine
    Data['Age_Pea']= Peace
    Data['Age_Phy']= Physics
    Data['Age_Eco']= Economics
    plt.rcParams['figure.figsize'] = [6,40]  
    sns.set_theme(style="darkgrid")
    fig, axs = plt.subplots(6, sharex=True, sharey=True)
    axs[0].scatter(Data['year'],Data['Age_che'],label='Low')
    sns.regplot(ax=axs[0],x=Data['year'], y= Data['Age_che'],label='low')    #Estimating regression fits 
    axs[0].set_ylabel('Age')
    axs[0].set_xlabel('')
    axs[0].set_title('Category = Chemistry',fontsize = 14)
    
    axs[1].scatter(Data['year'],Data['Age_Liter'],label='low')
    sns.regplot(ax=axs[1],x=Data['year'], y= Data['Age_Liter'],label='low')    #Estimating regression fits 
    axs[1].set_ylabel('Age')
    axs[1].set_xlabel('') 
    axs[1].set_title('Category = Literature',fontsize = 14)
    
    axs[2].scatter(Data['year'],Data['Age_Med'],label='low')
    sns.regplot(ax=axs[2],x=Data['year'], y= Data['Age_Med'],label='low')    #Estimating regression fits 
    axs[2].set_ylabel('Age')
    axs[2].set_xlabel('') 
    axs[2].set_title('Category = Medicine',fontsize = 14)
    
    axs[3].scatter(Data['year'],Data['Age_Pea'],label='low')
    sns.regplot(ax=axs[3],x=Data['year'], y= Data['Age_Pea'],label='low')    #Estimating regression fits 
    axs[3].set_ylabel('Age')
    axs[3].set_xlabel('') 
    axs[3].set_title('Category = Peace',fontsize = 14)
    
    axs[4].scatter(Data['year'],Data['Age_Phy'],label='low')
    sns.regplot(ax=axs[4],x=Data['year'], y= Data['Age_Phy'],label='low')    #Estimating regression fits 
    axs[4].set_ylabel('Age')
    axs[4].set_xlabel('')   
    axs[4].set_title('Category = Physics',fontsize = 14)
    
    axs[5].scatter(Data['year'],Data['Age_Eco'],label='low')
    sns.regplot(ax=axs[5],x=Data['year'], y= Data['Age_Eco'],label='low')    #Estimating regression fits 
    axs[5].set_ylabel('Age')
    axs[5].set_xlabel('') 
    axs[5].set_xticks(np.arange(1850, 2100, 50))
    axs[5].set_xticklabels([ '','1900', '1950', '2000','2050'])
    axs[5].set_title('Category = Economics')
    

    
#  sns.lmplot(
#         x='year',
#         y='Time',
#         data=Data,
#         row='category'
#         )
```
