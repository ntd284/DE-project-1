```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib.ticker import PercentFormatter

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()

with open('nobel.csv','r') as f:
    Data = pd.read_csv(f)
    Time = pd.to_datetime(Data["birth_date"]).dt.year
    Data['Age'] =  Data['year'] - Time   
    
    plt.rcParams['figure.figsize'] = [7,7] 
    
    plt.scatter(Data['year'], Data['Age'])
    
    sns.regplot(x=Data['year'], y= Data['Age'])    #Estimating regression fits 

    plt.grid()
    
    
        
#     sns.lmplot(
#         x = 'year',
#         y = 'Age',
#         data = Data,
#     )
```
